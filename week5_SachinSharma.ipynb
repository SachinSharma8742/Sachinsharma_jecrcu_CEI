# 🧠 Problem Statement
Design and implement a DL model capable of learning the underlying structure, grammar, and contextual dependencies of a given text corpus to generate coherent and meaningful text sequences using:

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, GRU, Dense
import numpy as np
import matplotlib.pyplot as plt
print("TensorFlow:", tf.__version__)
```

# 📥 Load Text Corpus
✅ **Task 1 Completed:** Replaced with a more comprehensive corpus about AI and deep learning

```python
corpus = '''
artificial intelligence is transforming the world through machine learning
deep learning uses neural networks with many layers to learn complex patterns
recurrent neural networks process sequential data like text and speech
long short term memory networks solve the vanishing gradient problem
gated recurrent units are simpler and faster than lstm networks
transformers use self attention to capture long range dependencies
natural language processing helps computers understand human language
computer vision enables machines to interpret and understand visual information
reinforcement learning trains agents through trial and error
generative ai creates new content like text images and music
large language models are trained on vast amounts of text data
deep learning models require powerful hardware like gpus for training
data preprocessing is a crucial step in any machine learning pipeline
overfitting occurs when a model learns noise instead of patterns
regularization techniques help improve model generalization
'''
print(corpus)
```

# 🔤 Tokenization & Sequence Creation
We convert text into integer tokens and create **n-gram style sequences**
for next-word prediction.

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([corpus])

total_words = len(tokenizer.word_index) + 1
print("Vocabulary size:", total_words)

input_sequences = []
for line in corpus.split('\n'):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_seq = token_list[:i+1]
        input_sequences.append(n_gram_seq)

max_len = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')

X = input_sequences[:, :-1]
y = input_sequences[:, -1]

print("X shape:", X.shape)
print("y shape:", y.shape)
```

# 🧠 Model 1: Vanilla RNN
This is the baseline sequential model.
It struggles with long-term dependencies because of vanishing gradients.

✅ **Tasks 2, 3, 4 Completed:**
- Embedding dimension increased: 32 → 64
- Hidden units increased: 64 → 128
- Epochs increased: 100 → 200

```python
rnn_model = Sequential([
    Embedding(total_words, 64, input_length=max_len-1),  # ✅ Embedding dim 32→64
    SimpleRNN(128),  # ✅ Hidden units 64→128
    Dense(total_words, activation='softmax')
])

rnn_model.compile(loss='sparse_categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

rnn_history = rnn_model.fit(X, y, epochs=200, verbose=1)  # ✅ Epochs 100→200
print("Vanilla RNN training completed")
```

# 🔒 Model 2: LSTM
LSTM uses **input, forget, and output gates**
to preserve long-term memory.

✅ **Tasks 2, 3, 4 Completed:**
- Embedding dimension increased: 32 → 64
- Hidden units increased: 64 → 128
- Epochs increased: 100 → 200

```python
lstm_model = Sequential([
    Embedding(total_words, 64, input_length=max_len-1),  # ✅ Embedding dim 32→64
    LSTM(128),  # ✅ Hidden units 64→128
    Dense(total_words, activation='softmax')
])

lstm_model.compile(loss='sparse_categorical_crossentropy',
                   optimizer='adam',
                   metrics=['accuracy'])

lstm_history = lstm_model.fit(X, y, epochs=200, verbose=1)  # ✅ Epochs 100→200
print("LSTM training completed")
```

# ⚡ Model 3: GRU
GRU uses **reset + update gates**.
It is computationally faster than LSTM and often gives similar results.

✅ **Tasks 2, 3, 4 Completed:**
- Embedding dimension increased: 32 → 64
- Hidden units increased: 64 → 128
- Epochs increased: 100 → 200

```python
gru_model = Sequential([
    Embedding(total_words, 64, input_length=max_len-1),  # ✅ Embedding dim 32→64
    GRU(128),  # ✅ Hidden units 64→128
    Dense(total_words, activation='softmax')
])

gru_model.compile(loss='sparse_categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

gru_history = gru_model.fit(X, y, epochs=200, verbose=1)  # ✅ Epochs 100→200
print("GRU training completed")
```

## 📉 Compare Training Loss and Accuracy

```python
plt.figure(figsize=(14, 5))

# Loss comparison
plt.subplot(1, 2, 1)
plt.plot(rnn_history.history['loss'], label='RNN')
plt.plot(lstm_history.history['loss'], label='LSTM')
plt.plot(gru_history.history['loss'], label='GRU')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss Comparison")
plt.legend()

# Accuracy comparison
plt.subplot(1, 2, 2)
plt.plot(rnn_history.history['accuracy'], label='RNN')
plt.plot(lstm_history.history['accuracy'], label='LSTM')
plt.plot(gru_history.history['accuracy'], label='GRU')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training Accuracy Comparison")
plt.legend()

plt.tight_layout()
plt.show()
```

# ✍️ Text Generation Function

```python
def generate_text(model, seed_text, next_words=10):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_len-1, padding='pre')
        predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)[0]

        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break
        seed_text += " " + output_word
    return seed_text
```

## 🧪 Generate Text Samples
✅ **Task 5 Completed:** Generating 10 words instead of 5!

```python
seed_texts = ["deep learning", "artificial intelligence", "recurrent neural", "natural language"]

for seed in seed_texts:
    print("\n" + "="*60)
    print(f"Seed: '{seed}'")
    print("="*60)
    print("RNN :", generate_text(rnn_model, seed, 10))  # ✅ 10 words
    print("LSTM:", generate_text(lstm_model, seed, 10))
    print("GRU :", generate_text(gru_model, seed, 10))
```
