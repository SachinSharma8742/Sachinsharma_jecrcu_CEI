### 1. The Calculator Tool
This simple tool evaluates a mathematical string. It handles basic exceptions gracefully.

```python

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
```

### 2. Keyword Extraction Tool
This function extracts meaningful words from a given text by filtering for length and uniqueness.

```python

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
```

### 3. Core Agent Pipeline
Here is the main routing logic. It checks for specific trigger words to route the query to the appropriate tool, and handles unrecognized queries or errors nicely.

```python
import json
import re
def agent(query: str):
    query_lower = query.lower()

    if "calculate" in query_lower:
        expression = re.sub(r'(?i)calculate', '', query).strip()
        result = calculator(expression)
        if result == "Error in calculation":
            return {
                "type": "error",
                "result": "Failed to parse or calculate expression. Bad format."
            }
        return {
            "type": "calculation",
            "result": result
        }

    elif "keywords" in query_lower:
        result = extract_keywords(query)
        if not result:
            return {
                "type": "error",
                "result": "Failed to extract keywords."
            }
        return {
            "type": "keywords",
            "result": result
        }

    else:
        return {
            "type": "general",
            "result": "I am a smart assistant. I can perform calculations or extract keywords. How can I help you?"
        }
```

### 4. Validation Checks
Let's run a few test cases through the agent to make sure the JSON outputs are correctly structured and errors are caught.

```python
import json
queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?",
    "Calculate fifty minus ten"
]

for q in queries:
    print("Query:", q)
    print("Response:", json.dumps(agent(q), indent=2))
    print("-" * 50)
```
